```python
#creating bind for age group
age_bins = [ 0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
bin_name = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#segmenting and sorting data into bins
purchase_data['Age'] = pd.cut(purchase_data['Age'], age_bins, labels=bin_name)

#creating data frame to be sorted into bins
age_stats = pd.DataFrame(purchase_data.groupby('Age').SN.nunique())

#renaming the index into string and renaming sn to player count
age_stats = age_stats.rename(index=str, columns={'SN': 'Player Count'})

#calculating the % of players
age_stats['Percentage of Players'] = age_stats['Player Count'] / sum(age_stats['Player Count'])

#formating the Percentage of players
age_stats['Percentage of Players'] = age_stats['Percentage of Players'].map("{:,.2%}".format)


age_stats
```

* Display a preview of the summary data frame



```python
#look for the top 5 spenders by grouping purchases by 'SN'
top_spend = pd.DataFrame(purchase_data.groupby('SN').agg({'Purchase ID':'count', 'Price': {'Average Purchase Price':'mean','Total Purchase Value':'sum' }}))

#collapse the multiple indexes to create a cleanly labeled data frame
top_spend.columns = top_spend.columns.get_level_values(1) 

#rename the columns to clarify what they are
top_spend = top_spend.rename(index=str, columns={'count': 'Purchase Count'})

#sort Total Purchase Value column in descending order to get the top 5 spenders
top_spend = top_spend.sort_values(by='Total Purchase Value', ascending=False)

#limit the the DataFrame so the 5 top rows are populated
top_spend = top_spend[0:5]

#convert formatting to currency
top_spend['Average Purchase Price'] = top_spend['Average Purchase Price'].map("${:,.2f}".format)
top_spend['Total Purchase Value'] = top_spend['Total Purchase Value'].map("${:,.2f}".format)
top_spend
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#track most popular items by number of purchases by grouping purchase data according to item id and name
data_items = purchase_data[['Item ID', 'Item Name', 'Price']]

#group the data by item number and name to summarize purchases accordingly
top_trans = pd.DataFrame(data_items.groupby(['Item ID', 'Item Name']).agg({'Price':['count', 'sum']}))

#colapse the column headers
top_trans.columns = top_trans.columns.get_level_values(1)

#add a new column that lists the original price
top_trans['Item Price'] = top_trans['sum']/top_trans['count']

#organize columns
top_trans = top_trans[['count', 'Item Price', 'sum']]

#rename columns to reflect the data they are conveying
top_trans = top_trans.rename(index=str, columns={'count': 'Purchase Count', 'sum': 'Total Purchase Value'})

#sort the DataFrame according to 'Purchase Count' to identify most popular items
top_trans = top_trans.sort_values(by='Purchase Count', ascending=False)

#display top 5 items
top5_trans = top_trans[0:5]

#convert formatting to currency
top5_trans['Item Price'] = top5_trans['Item Price'].map("${:,.2f}".format)
top5_trans['Total Purchase Value'] = top5_trans['Total Purchase Value'].map("${:,.2f}".format)

top5_trans
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#track most popular items by number of purchases by grouping purchase data according to item id and name
data_items = purchase_data[['Item ID', 'Item Name', 'Price']]

#group the data by item number and name to summarize purchases accordingly
top_trans = pd.DataFrame(data_items.groupby(['Item ID', 'Item Name']).agg({'Price':['count', 'sum']}))

#colapse the column headers
top_trans.columns = top_trans.columns.get_level_values(1)

#add a new column that lists the original price
top_trans['Item Price'] = top_trans['sum']/top_trans['count']

#order columns
top_trans = top_trans[['count', 'Item Price', 'sum']]

#rename columns to reflect the data they are conveying
top_trans = top_trans.rename(index=str, columns={'count': 'Purchase Count', 'sum': 'Total Purchase Value'})

#sort the DataFrame according to 'Purchase Count' to identify most popular items
top_trans = top_trans.sort_values(by='Total Purchase Value', ascending=False)

#show top 5 item
top5_trans = top_trans[0:5]

#currency format
top5_trans['Item Price'] = top5_trans['Item Price'].map("${:,.2f}".format)
top5_trans['Total Purchase Value'] = top5_trans['Total Purchase Value'].map("${:,.2f}".format)

top5_trans
```
